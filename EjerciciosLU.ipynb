```python
import numpy as np

# Datos muestrales
datos = [10, 8, 11, 7, 9]

print("Datos:", datos)
print()

media = np.mean(datos)
print(f"Media (μ): {media}")

desviacion_estandar = np.std(datos, ddof=1) #ddof=1 es para que sea desv est. muestral
print(f"Desviación estándar muestral (s): {desviacion_estandar:.4f}") # .4f para que tenga solo 4 cifras luego de la coma

coeficiente_variacion = (desviacion_estandar / media) * 100
print(f"Coeficiente de variación (CV): {coeficiente_variacion:.2f}%")
```

```python
import numpy as np
import pandas as pd

clases = ['0-4', '5-9', '10-14', '15-19', '20-24']
frecuencias = [5, 8, 11, 9, 7]

# Calculamos los puntos medios a mano --> (0+4)/2, (5+9)/2, etc
marcas_clase = [2, 7, 12, 17, 22]  

n = sum(frecuencias)

# a) Calcular la media muestral
# Media = Σ(xi * fi) / n
suma_xi_fi = sum(xi * fi for xi, fi in zip(marcas_clase, frecuencias))
media = suma_xi_fi / n

# b) Calcular la varianza y desviación estándar muestral
# Varianza muestral = Σ(xi - x̄)² × fi / (n-1)
# Primero calculamos (xi - x̄)² × fi para cada clase
diferencias_cuadrado = [(xi - media)**2 for xi in marcas_clase]
diferencias_cuadrado_fi = [(xi - media)**2 * fi for xi, fi in zip(marcas_clase, frecuencias)]

# Suma de (xi - x̄)² × fi
suma_diferencias_cuadrado_fi = sum(diferencias_cuadrado_fi)

# Varianza muestral (dividimos por n-1)
varianza_muestral = suma_diferencias_cuadrado_fi / (n - 1)

# Desviación estándar muestral
desviacion_estandar_muestral = np.sqrt(varianza_muestral)

print(f"a) Media muestral: {media:.2f}")
print(f"b) Varianza muestral: {varianza_muestral:.4f}")
print(f"   Desviación estándar muestral: {desviacion_estandar_muestral:.4f}")
```

```python
import numpy as np
import pandas as pd

defectos = [0, 1, 2, 3]  # xi (número de defectos)
num_radios = [12, 15, 17, 6]  # fi (frecuencias)

n = sum(num_radios)

# Paso 1: Calcular la media
suma_xi_fi = sum(xi * fi for xi, fi in zip(defectos, num_radios))
media = suma_xi_fi / n

# Paso 2: Calcular la varianza muestral
suma_diferencias_cuadrado_fi = 0
for i in range(len(defectos)):
    xi = defectos[i]
    fi = num_radios[i]
    diferencia = xi - media
    diferencia_cuadrado = diferencia ** 2
    diferencia_cuadrado_fi = diferencia_cuadrado * fi
    suma_diferencias_cuadrado_fi += diferencia_cuadrado_fi

# Varianza muestral
varianza_muestral = suma_diferencias_cuadrado_fi / (n - 1)

# Paso 3: Calcular la desviación estándar
desviacion_estandar = np.sqrt(varianza_muestral)
print(f"Desviación estándar = {desviacion_estandar:.4f}")
```

```python
import numpy as np
import pandas as pd

x = [1, 3, 4, 3, 5, 5]
y = [5, 7, 6, 6, 7, 8]
n = len(x)

# Calcular medias
media_x = np.mean(x)
media_y = np.mean(y)

# a) Calcular la covarianza
suma_productos = 0
for i in range(n):
    diff_x = x[i] - media_x
    diff_y = y[i] - media_y
    producto = diff_x * diff_y
    suma_productos += producto

covarianza = suma_productos / (n - 1)

# b) Calcular el coeficiente de correlación
# Calcular desviaciones estándar
sx = np.std(x, ddof=1)  
sy = np.std(y, ddof=1)  

suma_cuadrados_x = 0
for i in range(n):
    diff_x = x[i] - media_x
    cuadrado_x = diff_x ** 2
    suma_cuadrados_x += cuadrado_x

suma_cuadrados_y = 0
for i in range(n):
    diff_y = y[i] - media_y
    cuadrado_y = diff_y ** 2
    suma_cuadrados_y += cuadrado_y

# Coeficiente de correlación
correlacion = covarianza / (sx * sy)

print(f"a) Covarianza muestral: {covarianza:.4f}")
print(f"b) Coeficiente de correlación: {correlacion:.4f}")

# Interpretación del coeficiente de correlación
print("\nInterpretación del coeficiente de correlación:")
if abs(correlacion) >= 0.8:
    fuerza = "fuerte"
elif abs(correlacion) >= 0.5:
    fuerza = "moderada"
else:
    fuerza = "débil"

if correlacion > 0:
    direccion = "positiva"
elif correlacion < 0:
    direccion = "negativa"
else:
    direccion = "nula"

print(f"r = {correlacion:.4f} indica una correlación {fuerza} {direccion}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos: (costo, días) - X = costo, Y = días
datos = [(3,7), (3,6), (3,5), (5,5), (10,2), (3,9), (5,6), (10,5), 
         (5,6), (10,1), (3,10), (5,7), (10,4), (5,6), (10,4)]

x = [punto[0] for punto in datos]  # Costo de envío ($)
y = [punto[1] for punto in datos]  # Días de envío

print("DATOS DEL SERVICIO ACME DELIVERY")
print("=" * 50)
print("Servicios disponibles:")
print("- Ordinario: $3")
print("- Urgente: $5") 
print("- Superurgente: $10")
print()

# Crear DataFrame para mejor visualización
df = pd.DataFrame({
    'Envío': range(1, len(datos)+1),
    'Costo ($)': x,
    'Días': y
})

print("Datos completos:")
print(df)
print()

n = len(x)
print(f"Total de envíos: {n}")

# Estadísticas descriptivas por servicio
print("\nEstadísticas por tipo de servicio:")
print("-" * 40)

for costo in [3, 5, 10]:
    dias_servicio = [y[i] for i in range(len(x)) if x[i] == costo]
    if costo == 3:
        nombre = "Ordinario ($3)"
    elif costo == 5:
        nombre = "Urgente ($5)"
    else:
        nombre = "Superurgente ($10)"
    
    print(f"{nombre}:")
    print(f"  Días promedio: {np.mean(dias_servicio):.1f}")
    print(f"  Días mín-máx: {min(dias_servicio)}-{max(dias_servicio)}")
    print(f"  Número de envíos: {len(dias_servicio)}")
    print()

print("=" * 50)
```

```python
# a) DESCRIPCIÓN NUMÉRICA: COVARIANZA Y CORRELACIÓN
print("a) ANÁLISIS NUMÉRICO:")
print("=" * 30)

# Calcular medias
media_x = np.mean(x)  # Media del costo
media_y = np.mean(y)  # Media de los días

print(f"Media del costo (x̄): ${media_x:.2f}")
print(f"Media de los días (ȳ): {media_y:.2f} días")
print()

# COVARIANZA
print("COVARIANZA:")
print("Covarianza = Σ[(xi - x̄)(yi - ȳ)] / (n-1)")
print()

# Cálculos detallados
print("Cálculos detallados:")
print("i  | xi | yi | (xi-x̄) | (yi-ȳ) | (xi-x̄)(yi-ȳ)")
print("---|----|----|--------|--------|--------------")

suma_productos = 0
for i in range(n):
    diff_x = x[i] - media_x
    diff_y = y[i] - media_y
    producto = diff_x * diff_y
    suma_productos += producto
    print(f"{i+1:2d} | {x[i]:2d} | {y[i]:2d} | {diff_x:6.2f} | {diff_y:6.2f} | {producto:8.2f}")

covarianza = suma_productos / (n - 1)
print(f"Σ[(xi-x̄)(yi-ȳ)] = {suma_productos:.2f}")
print(f"Covarianza = {suma_productos:.2f} / {n-1} = {covarianza:.4f}")
print()

# COEFICIENTE DE CORRELACIÓN
print("COEFICIENTE DE CORRELACIÓN:")
print("r = Covarianza / (sx × sy)")
print()

# Desviaciones estándar
sx = np.std(x, ddof=1)
sy = np.std(y, ddof=1)

print(f"Desviación estándar del costo (sx): {sx:.4f}")
print(f"Desviación estándar de los días (sy): {sy:.4f}")

correlacion = covarianza / (sx * sy)
print(f"Coeficiente de correlación (r): {covarianza:.4f} / ({sx:.4f} × {sy:.4f})")
print(f"r = {correlacion:.4f}")
print()

# Interpretación
print("INTERPRETACIÓN:")
print(f"• Covarianza = {covarianza:.4f}")
if covarianza < 0:
    print("  → Relación NEGATIVA: a mayor costo, menor tiempo de entrega")
elif covarianza > 0:
    print("  → Relación POSITIVA: a mayor costo, mayor tiempo de entrega")
else:
    print("  → No hay relación lineal")

print(f"• Correlación = {correlacion:.4f}")
if abs(correlacion) >= 0.8:
    fuerza = "FUERTE"
elif abs(correlacion) >= 0.5:
    fuerza = "MODERADA"
elif abs(correlacion) >= 0.3:
    fuerza = "DÉBIL"
else:
    fuerza = "MUY DÉBIL"

if correlacion < 0:
    direccion = "NEGATIVA"
elif correlacion > 0:
    direccion = "POSITIVA"
else:
    direccion = "NULA"

print(f"  → Correlación {fuerza} {direccion}")
print(f"  → {abs(correlacion)*100:.1f}% de la variación en días se explica por el costo")
print()

print("=" * 60)
```

```python
# b) ANÁLISIS DEL VALOR DE LOS SERVICIOS MÁS CAROS
print("b) ANÁLISIS DEL VALOR DE LOS SERVICIOS:")
print("=" * 40)

# Análisis detallado por servicio
servicios = {
    3: {"nombre": "Ordinario", "dias": []},
    5: {"nombre": "Urgente", "dias": []},
    10: {"nombre": "Superurgente", "dias": []}
}

# Separar datos por servicio
for i in range(n):
    servicios[x[i]]["dias"].append(y[i])

# Calcular estadísticas por servicio
for costo, info in servicios.items():
    dias = info["dias"]
    info["promedio"] = np.mean(dias)
    info["min"] = min(dias)
    info["max"] = max(dias)
    info["desv_std"] = np.std(dias, ddof=1)
    info["envios"] = len(dias)

print("RESUMEN COMPARATIVO:")
print("Servicio      | Costo | Días Prom | Rango    | Desv.Est | Envíos")
print("--------------|-------|-----------|----------|----------|--------")
for costo in [3, 5, 10]:
    info = servicios[costo]
    print(f"{info['nombre']:12s} | ${costo:4d} | {info['promedio']:8.1f} | {info['min']}-{info['max']:7s} | {info['desv_std']:7.2f} | {info['envios']:6d}")

print()

# Análisis de eficiencia (costo por día ahorrado)
print("ANÁLISIS DE EFICIENCIA:")
print("-" * 25)

ordinario_prom = servicios[3]["promedio"]
urgente_prom = servicios[5]["promedio"]
superurgente_prom = servicios[10]["promedio"]

# Comparar Urgente vs Ordinario
ahorro_urgente = ordinario_prom - urgente_prom
costo_extra_urgente = 5 - 3
if ahorro_urgente > 0:
    eficiencia_urgente = costo_extra_urgente / ahorro_urgente
    print(f"URGENTE vs ORDINARIO:")
    print(f"• Costo extra: ${costo_extra_urgente}")
    print(f"• Días ahorrados: {ahorro_urgente:.1f}")
    print(f"• Costo por día ahorrado: ${eficiencia_urgente:.2f}/día")
else:
    print(f"URGENTE vs ORDINARIO: No hay ahorro significativo")

print()

# Comparar Superurgente vs Ordinario
ahorro_super = ordinario_prom - superurgente_prom
costo_extra_super = 10 - 3
if ahorro_super > 0:
    eficiencia_super = costo_extra_super / ahorro_super
    print(f"SUPERURGENTE vs ORDINARIO:")
    print(f"• Costo extra: ${costo_extra_super}")
    print(f"• Días ahorrados: {ahorro_super:.1f}")
    print(f"• Costo por día ahorrado: ${eficiencia_super:.2f}/día")
else:
    print(f"SUPERURGENTE vs ORDINARIO: No hay ahorro significativo")

print()

# Comparar Superurgente vs Urgente
ahorro_super_vs_urgente = urgente_prom - superurgente_prom
costo_extra_super_vs_urgente = 10 - 5
if ahorro_super_vs_urgente > 0:
    eficiencia_super_vs_urgente = costo_extra_super_vs_urgente / ahorro_super_vs_urgente
    print(f"SUPERURGENTE vs URGENTE:")
    print(f"• Costo extra: ${costo_extra_super_vs_urgente}")
    print(f"• Días ahorrados: {ahorro_super_vs_urgente:.1f}")
    print(f"• Costo por día ahorrado: ${eficiencia_super_vs_urgente:.2f}/día")
else:
    print(f"SUPERURGENTE vs URGENTE: No hay ahorro significativo")

print()
print("CONCLUSIONES:")
print("=" * 15)
print("1. RELACIÓN COSTO-TIEMPO:")
if correlacion < -0.5:
    print("   ✓ Existe una relación negativa significativa entre costo y tiempo")
    print("   ✓ Los servicios más caros SÍ entregan más rápido")
else:
    print("   ✗ La relación entre costo y tiempo no es muy clara")

print()
print("2. VALOR DE LOS SERVICIOS PREMIUM:")

# Evaluar si vale la pena
if ahorro_urgente > 1:
    print(f"   ✓ Urgente: Vale la pena (ahorra {ahorro_urgente:.1f} días por $2 extra)")
else:
    print(f"   ✗ Urgente: Beneficio marginal ({ahorro_urgente:.1f} días por $2 extra)")

if ahorro_super > 2:
    print(f"   ✓ Superurgente: Vale la pena (ahorra {ahorro_super:.1f} días por $7 extra)")
elif ahorro_super > 1:
    print(f"   ? Superurgente: Beneficio moderado ({ahorro_super:.1f} días por $7 extra)")
else:
    print(f"   ✗ Superurgente: Beneficio limitado ({ahorro_super:.1f} días por $7 extra)")

print()
print("3. RECOMENDACIÓN:")
if eficiencia_urgente < eficiencia_super:
    print("   → El servicio URGENTE ofrece mejor relación costo-beneficio")
else:
    print("   → El servicio SUPERURGENTE puede justificarse para envíos críticos")
```
